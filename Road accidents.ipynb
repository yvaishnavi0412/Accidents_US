# Road Accidents Data Analysis 

This project aims to find out the different factors that contribute to a severe road accident

# Data

The dataset encompasses car accidents nationwide across all 49 states of the USA, spanning from February 2016 to June 2023. It is compiled using two APIs that stream traffic incident data sourced from multiple entities including transportation departments, law enforcement agencies, traffic cameras, and sensors within road networks. Presently, the dataset comprises approximately 3.5 million accident records.

# Key Findings

The analysis reveals a concentration of accidents in Miami, FL, particularly during daylight hours in December. Furthermore, significant accidents predominantly occur between 3 pm to 5 pm. Notably, the data indicates a notable increase in accidents during the year 2022.

# What time of the year is the percent of accident high?

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'],format='%Y-%m-%d %H:%M:%S',errors='coerce')
```

sns.histplot(df['Start_Time'].dt.month, color='orange', bins=12, stat='percent');

```python
mi=df[df.City=='Miami']
group_mi=mi.groupby(mi['Start_Time'].dt.month).agg({'Severity':'sum'})
months=group_mi.index
severity_values=group_mi['Severity']
plt.bar(months, severity_values,color='lightblue')
plt.xlabel('Months')
plt.ylabel('Total Severity')
plt.xticks(months)
plt.title('Accidents in Miami by Month')

plt.show()
```

# Period of day most accidents occurred

```python
state_lst=['FL','GA','TX']
state_lst_full=['Florida','Georgia','Texas']

feature='Sunrise_Sunset'

fig= plt.figure(figsize=(15,6))

pct_cutoff=2


def my_autopct(pct):
    return ('%1.0f%%' % pct) if pct > pct_cutoff else ''



for i,state in enumerate(state_lst):
    plt.subplot(1, 3, 1+i)
    #Slice the dataframe for the specific state and feature
    df_temp=df[df['State']==state][feature].value_counts(normalize=True).round(2)
    #Define lables to go with the pie plot
    labels = [n if v > pct_cutoff/100 else ''
              for n, v in zip(df_temp.index, df_temp)] 
    
    plt.pie(df_temp, labels=labels, autopct=my_autopct, shadow=True)
    plt.axis('equal')
    plt.xlabel(feature)
    plt.title(state)
    
plt.tight_layout()
plt.show()
```
